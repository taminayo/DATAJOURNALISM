```python
from bs4 import BeautifulSoup
import requests
after = '/tag/%EB%B2%9A%EA%BD%83%EC%97%94%EB%94%A9?cursor=J0HWjURCQAAAF0HWhe2AgAAAFsgDAA%253D%253D'
f = open('please.text', 'a')
url = 'https://web.stagram.com' + after
soup
text = ''
days = []
texts = []
nofp = 0
noft = 0
realdays = []
def mkrequest():
    global request
    global text
    global soup
    try:
        request = requests.get(url)
    except:
        print("request error")
        return
    text = request.text
    soup = BeautifulSoup(text, 'html.parser')

def mkdays():
    global days
    global soup
    print(soup.find_all(class_ = 'post_date')[0])
    for element in soup.find_all(class_ = 'post_date'):
        global days
        days.append(element.get_text())
    

def mktexts():
    global soup
    global nofp
    global noft
    global realdays
    arr = soup.find_all('a')
    gostr = ''
    for i in range(0,len(arr)):
        target = arr[i].get_text()
        if '#' in target:
            gostr = gostr+target
        elif not gostr == '':
            if '버스커' in gostr and '벚꽃엔딩' in gostr:
                texts.append(gostr)
                nofp = nofp + 1
                gostr = ''
            else:
                gostr = ''

def gonext():
    global after
    global soup
    try:
        after = soup.find(class_ = 'btn btn-primary')['href']
    except:
        print("after error")
        return


def main():
    global url
    global after
    global soup
    global days
    global texts
    global ff
    global realdays
    url = 'https://web.stagram.com' + after
    days = []
    texts = []
    mkrequest()
    mkdays()
    mktexts()
    gonext()
    f.write(days[-1]+'\n')
    for i in range(0,len(texts)):
        f.write(texts[i]+'\n')
    f.close()
    print(days[0])
    print(after)
    if nofp < 10:
        try:
            main()
        except:
            print("recurrsion error")
            return


main()
```
